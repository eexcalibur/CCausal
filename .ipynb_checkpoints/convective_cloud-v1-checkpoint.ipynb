```python
import pandas as pd
import sys
from pathlib import Path
home = str(Path.home())
sys.path.append(home+"/climate_causal/utils/")
import algs
import PostPF
import lingam
import numpy as np
from graph_tool.all import *
import matplotlib.pyplot as plt
#plt.switch_backend('cairo')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline 
```

# load index data

```python
var_names = ['mucape [J/kg]','low_level_cape [J/kg]','mucin [J/kg]', 'lnb [km]', 
             'lfc [km]','lcl [km]','low_level_wind_shear [m/s]','low_level_rh [%]',
             'elr_3km [C/km]','elr_6km [C/km]','lmd_mean [km]']

var_sh = ['MCAPE','LCAPE','CIN','LNB','LFC','LCL','WSH','RH','ELR3','ELR6','LMD']

def get_data(f_data):
    data = pd.read_csv('outputs/oceanic_arm_tropical_detrainment_data_mu_ice.csv')
    print(f'before nan: {data.shape}')
    data.dropna(inplace=True)
    print(f'after nan: {data.shape}')
    data = data[data['low_level_wind_shear [m/s]'] < 1000]
    print(f'after filter for wind shear: {data.shape}')
    
    data1 = data[var_names]
    data1.columns = var_sh
    
    return data1
```

```python
#f_data = pd.read_csv('outputs/twp_arm_tropical_detrainment_data_mu_ice.csv')
f_data = pd.read_csv('outputs/oceanic_arm_tropical_detrainment_data_mu_ice.csv')
#f_data = pd.read_csv('outputs/continental_arm_tropical_detrainment_data_mu_ice.csv')
#f_data = pd.read_csv('outputs/all_arm_tropical_detrainment_data_mu_ice.csv')
data = get_data(f_data)
```

# causal method

```python
model = lingam.ICALiNGAM()
model.fit(data)
x_adj = pd.DataFrame(np.transpose(model.adjacency_matrix_), columns = var_sh)

PostPF.create_graph_by_adjmatrix(x_adj, pos_name="cycle", show_weight=True)
```
